```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Cargar la imagen
img = cv2.imread("luna_JPG.jpg")

# Mostrar la imagen original
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Imagen original")
plt.show()

print("Shape:", img.shape)
```

```python
# Convertir a escala de grises promediando los 3 canales
X = (img[:, :, 0] + img[:, :, 1] + img[:, :, 2]) / 3

plt.imshow(X, cmap="gray")
plt.title("Escala de grises")
plt.show()

print("Valor mínimo:", np.min(X))
print("Valor máximo:", np.max(X))
```

```python
def segmenta(X, t):
    N = X.shape[0]  # filas
    M = X.shape[1]  # columnas

    # Ayuda 2: inicializar con 255 en vez de 0
    Y = np.ones((N, M)) * 255

    for i in range(N):
        for j in range(M):
            if X[i, j] > t:
                Y[i, j] = 255  # píxel de la luna → blanco
            else:
                Y[i, j] = 0    # fondo negro → 0

    return Y
```

```python
t = 15  # <-- Probar modificando este valor


resultado = segmenta(X, t)


# Ayuda 4: usar cmap="gray"
plt.imshow(resultado, cmap="gray")
plt.title(f"Segmentación con umbral t={t}")
plt.show()
```
